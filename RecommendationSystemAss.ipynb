```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book = pd.read_csv('/content/book (1).csv',encoding='latin-1')
book.head()
```

# **EDA**

```python
book1 = book.drop(book.columns[[0]], axis=1)
```

```python
book1.head()
```

```python
book1.rename(columns = {'User.ID':'UserID','Book.Title':'BookTittle','Book.Rating':'Book_Rating'}, inplace = True)
```

```python
book1
```

```python
len(book1.UserID.unique())
```

```python
len(book1.BookTittle.unique())
```

```python
book1 =book1.sort_values(by=['UserID'])
```

```python
book2 = book1.pivot_table(index='UserID',
                   columns='BookTittle',
                   values='Book_Rating').reset_index(drop=True)
```

```python
book2
```

```python
book2.fillna(0,inplace=True)
```

```python
book2
```

# **Cosine Similarity between Users**

```python
user_sim = 1 - pairwise_distances(book2.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = book1.UserID.unique()
user_sim_df.columns = book1.UserID.unique()
```

```python
user_sim_df.iloc[0:10,0:10]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
#Most Similer User
print(user_sim_df.idxmax(axis=1)[12])
print(user_sim_df.max(axis=1).sort_values(ascending=False).head(20))
```
